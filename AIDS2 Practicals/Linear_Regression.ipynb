```python
# Predicting Patient Recovery Time using Linear Regression

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error

# Step 1: Example dataset (you can replace this with your patient_data.csv)
data = {
    'Age': [25, 30, 35, 40, 45, 50, 55, 60],
    'Treatment_Duration': [5, 7, 8, 10, 12, 14, 15, 18],
    'Recovery_Time': [6, 8, 9, 11, 13, 15, 16, 19]  # Target
}
df = pd.DataFrame(data)

# Step 2: Preprocessing
X = df[['Age', 'Treatment_Duration']]
y = df['Recovery_Time']

# Step 3: Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 4: Train model
model = LinearRegression()
model.fit(X_train, y_train)

# Step 5: Predict
y_pred = model.predict(X_test)

# Step 6: Evaluate
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print("Predicted Recovery Times:", y_pred)
print("R² Score:", r2)
print("Mean Absolute Error:", mae)
```

```python
# Predicting Patient Recovery Time using Linear Regression

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error

# Step 1: Read dataset (make sure patient_data.csv is in same folder)
df = pd.read_csv("patient_data.csv")

# Step 2: Select features and target column
X = df[['Age', 'Treatment_Duration']]
y = df['Recovery_Time']

# Step 3: Split dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 4: Train Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Step 5: Predict
y_pred = model.predict(X_test)

# Step 6: Evaluate model
print("R² Score:", r2_score(y_test, y_pred))
print("Mean Absolute Error:", mean_absolute_error(y_test, y_pred))
```
